```python
import numpy as np
import pandas as pd
```

```python
cou_city_dtype = {
    'id': np.int64,
    'lat': str,
    'lng': str
}
```

```python
df_cou = pd.read_csv('./formatted_county_coord.csv', encoding='utf-8', dtype=cou_city_dtype)
df_rest = pd.read_csv('./cou_city_loc_rest1.csv', encoding='utf-8')
```

```python
df_rest_loc = pd.merge(df_rest, df_cou, how='inner', on='id')
```

```python
df_rest_loc
```

```python
df_rest_loc.to_csv('cou_rest_loc.csv', columns=['id', 'address', 'region', 'lat', 'lng'], encoding='utf-8', index=False)
```
